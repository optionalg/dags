# 무신사 기호만 제거한 버전

```python
import csv
import re
```

```python
musinsa_komoran_csv = []
# 품사 태깅이 기호(S로 시작)인 것은 모두 제거
for review in musinsa_review:
    musinsa_komoran = []
    result = komoran.get_plain_text(review)
    for word in result.split(' '):
        if not re.search('^S', word.split('/')[1]):
            tmp = re.sub('[(ㄱ-ㅎ)|(ㅏ-ㅣ)|(-=+,#/\?:^$.@*\"※~&%ㆍ!)]','',word.split('/')[0])
            musinsa_komoran.append(tmp)
    musinsa_komoran_csv.append(musinsa_komoran)
    
filename = './musinsa_구두_komoran_simple.csv'
f = open(filename, 'w', encoding='utf-8', newline='')
csvwriter = csv.writer(f)
for review in musinsa_komoran_csv:
    csvwriter.writerow(review)
f.close()
```
